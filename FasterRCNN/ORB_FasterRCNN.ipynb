```python
import cv2
import numpy as np

img = cv2.imread('car_2.jpg', 1)
```

```python
orb = cv2.ORB_create(nfeatures = 10, nlevels = 8, scaleFactor = 1.2)

keypoints, descriptors = orb.detectAndCompute(img, None)
imgg = cv2.drawKeypoints(img, keypoints, None, color=(0,255,0))

img = cv2.resize(img, (500,500))
imgg = cv2.resize(imgg, (500,500))
```

```python
from google.colab.patches import cv2_imshow

#cv2_imshow(img)
cv2_imshow(imgg)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from matplotlib import pyplot as plt
import gluoncv as gcv
from gluoncv import model_zoo, data, utils
import mxnet as mx
```

```python
net = model_zoo.get_model('faster_rcnn_resnet50_v1b_voc', pretrained=True)
```

```python
im_fname = 'car_2.jpg'
```

```python
x_img, orig_img = data.transforms.presets.rcnn.load_test(im_fname)
```

```python
box_ids, scores, bboxes = net(x_img)
ax = utils.viz.plot_bbox(orig_img, bboxes[0], scores[0], box_ids[0], class_names=net.classes)

plt.show()
```
